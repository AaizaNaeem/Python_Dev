```python
# This is the code to check the duplicate rows in original CSV File

import pandas as pd

input_file = '/content/flow_features.csv'  # Replace with your input CSV file

# Read the CSV file into a DataFrame
df = pd.read_csv(input_file)

# Find and print duplicate rows based on all columns
duplicate_rows = df[df.duplicated(keep=False)]
print("Duplicate Rows:")
print(duplicate_rows)
```

```python
# This is the code to remove the duplicates rows from the code

import pandas as pd

input_file = '/content/flow_features.csv'  # Replace with your input CSV file
output_file = '/content/flow_features.csv'  # Replace with your output CSV file

# Read the CSV file into a DataFrame
df = pd.read_csv(input_file)

# Remove duplicate rows based on all columns
df = df.drop_duplicates()

# Write the filtered DataFrame to a new CSV file
df.to_csv(output_file, index=False)
```
